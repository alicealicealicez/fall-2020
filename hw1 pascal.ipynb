```python
class binomial_coefficients:
    def __init__(self):
        pass
        
    def get_n(self, n):
        self.temp = []
        if n == 1:
            self.line = [1]
        elif n == 2:
            self.line = [1,1]
        else:
            self.line = [1,1]
            for i in range(n-2):
                self.temp = []
                self.temp.append(1)
                for j in range(i+1):
                    self.temp.append(self.line[j]+self.line[j+1])
                self.temp.append(1)
                self.line = self.temp
        print (self.line)
                
    def get_nk(self, n, k):
        self.temp = []
        if n == 1:
            self.line = [1]
        elif n == 2:
            self.line = [1,1]
        else:
            self.line = [1,1]
            for i in range(n-2):
                self.temp = []
                self.temp.append(1)
                for j in range(i+1):
                    self.temp.append(self.line[j]+self.line[j+1])
                self.temp.append(1)
                self.line = self.temp
        print (self.line[k])
    
    def save_pt(self, n, file_name = 'pascal_triangle.txt'):
        with open (file_name, 'wt') as f:
            self.temp = []
            if n == 1:
                self.line = [1]
            elif n == 2:
                self.line = [1,1]
            else:
                self.line = [1,1]
                for i in range(n-2):
                    self.temp = []
                    self.temp.append(1)
                    for j in range(i+1):
                        self.temp.append(self.line[j]+self.line[j+1])
                    self.temp.append(1)
                    self.line = self.temp
            llist = self.line
            self.temp = []
            line0 = [1]
            line1 = [1,1]
            if n == 1:
                line = ' '.join(map(str, line0))
                f.write(f'{line :^{n*2+1}}')
            elif n == 2:
                line = ' '.join(map(str, line0))
                f.write(f'{line :^{n*2+1}}'+'\n')
                line = ' '.join(map(str, line1))
                f.write(f'{line :^{n*2+1}}'+'\n')
            else:
                last = ' '.join(map(str, llist))
                num = len(last)
                line = ' '.join(map(str, line0))
                f.write(f'{line :^{num}}'+'\n')
                line = ' '.join(map(str, line1))
                f.write(f'{line :^{num}}'+'\n')
                self.line = [1,1]
                for i in range(n-2):
                    self.temp = []
                    self.temp.append(1)
                    for j in range(i+1):
                        self.temp.append(self.line[j]+self.line[j+1])
                    self.temp.append(1)
                    self.line = self.temp
                    line = ' '.join(map(str, self.line))
                    f.write(f'{line :^{num}}'+'\n')
        f.close()

        
```
