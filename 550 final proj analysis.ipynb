```python
from sklearn.preprocessing import StandardScaler
features = ['Tyrosine', 'Tyramine', '4-hydroxyphenyllactic acid', '4-hydroxyphenylacetic acid', '4-hydroxyphenyl propionic acid',
       'Phenylalanine', 'Phenyllactic acid', 'Phenylacetic acid',
       'Phenylpropionic acid', 'Tryptophan', 'Tryptamine', 'Kynurenine',
       'Kynurenic acid', 'Xanthurenic acid', 'Indolelactic acid',
       'Indoleacetic acid', 'Indolepropionic acid', 'Indolealdehyde',
       'Indoleethanol']
# Separating out the features
x = df.loc[:, features].values
# Separating out the target
#y = df.loc[:,['target']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
df1 = PCA(n_components=2)
principalComponents = df1.fit_transform(x)
```
